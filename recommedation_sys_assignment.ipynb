```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
books = pd.read_csv('/content/book.csv',encoding='Latin1')
books
```

```python
data = books.iloc[:,1:]
data
```

```python
# sort by user id
data.sort_values(['User.ID'])
```

```python
# number of unique books
len(data['Book.Title'].unique())
```

```python
 user_books= data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
user_books
```

```python
# replacing index values by unique user id's
user_books.index = data['User.ID'].unique()
user_books.index
```

```python
user_books
```

```python
# impute those NaNs with 0 values
user_books.fillna(0,inplace=True)
```

```python
user_books
```

```python
# calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1-pairwise_distances(user_books.values,metric='cosine')
```

```python
user_sim
```

```python
# store the results in the dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
# set the index and column names to user id
user_sim_df.index= data['User.ID'].unique()
user_sim_df.columns = data['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
# most similar user
user_sim_df.idxmax(axis=1)
```
